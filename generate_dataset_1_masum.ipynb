```python
# https://github.com/c2nes/javalang/blob/master/javalang/tokenizer.py
VALUES = ['>>>=', '>>=', '<<=',  '%=', '^=', '|=', '&=', '/=',
                  '*=', '-=', '+=', '<<', '--', '++', '||', '&&', '!=',
                  '>=', '<=', '==', '%', '^', '|', '&', '/', '*', '-',
                  '+', ':', '?', '~', '!', '<', '>', '=', '...', '->', '::', '\\', '\\\\', '\*', '*\\', '\\\\\\']
INFIX = ['||', '&&', '|', '^', '&', '==', '!=', '<', '>', '<=', '>=', 
             '<<', '>>', '>>>', '+', '-', '*', '/', '%']
PREFIX = ['++', '--', '!', '~', '+', '-']
POSTFIX = ['++', '--']
ASSIGNMENT = ['=', '+=', '-=', '*=', '/=', '&=', '|=', '^=', '%=', '<<=', '>>=', '>>>=']
LAMBDA = ['->']
COMMENT = ['//', '/*', '*/']
METHOD_REFERENCE = ['::']

WHITESPACES = ['<|xvis|>','<|xiis|>','<|viis|>','<|ivs|>','<|iis|>','<|is|>','<|it|>','<|iit|>','<|iiit|>','<|ivt|>','<|nl|>']
WHITESPACE_DICT = {"                ":'<|xvis|>', 
                   "            ":'<|xiis|>', 
                   "        ":'<|viiis|>', 
                   "    ":'<|ivs|>',
                   "  ":'<|iis|>'," ":'<|is|>',
                   "\t\t\t\t":'<|ivt|>',"\t\t\t":'<|iiit|>',"\t\t":'<|iit|>',"\t":'<|it|>',"\n":'<|nl|>'}

CUSTOM_TOKEN = ["<|startcode|>", "<|endcode|>", "<|startfocus|>", "<|endfocus|>", "<|startcomment|>", "<|endcomment|>", 
                   "<|stringliteral|>", "<|singlelinecomment|>", "<|multilinecomment|>", "<|del|>"]
```

```python
word_punct_tokenizer = WordPunctTokenizer()
tweet_tokenizer = TweetTokenizer()
mwe_tokenizer = MWETokenizer(token_phrases, separator="")

import re
## we might need this >> https://www.tutorialspoint.com/python/python_reg_expressions.htm


def getCommentToken(comment, max_size=MAX_COMMENT_TOKEN_STREAM_SIZE, code_like_tokenize=False):
    ## remove NULL character ('\u0000') from the comment text
#     comment = comment.replace('\u0000', '')
    ## for replacing all non-ascii characters: https://stackoverflow.com/questions/40872126/python-replace-non-ascii-character-in-string/40872225
    ## for non-ascii removal: comment=re.sub(r'[^\x00-\x7f]',r'', comment)
    ## for non-ascii+ascii-control-characters removal:  comment=re.sub(r'[^\x20-\x7E]',r'', comment)
    
    if code_like_tokenize:
        ## for string-literal: https://stackoverflow.com/questions/2039795/regular-expression-for-a-string-literal-in-flex-lex
        comment = re.sub(r'["]([^\\\"]|\\.)*["]', r"<STRING_LITERAL>", comment) # replace all string literals with <STRING_LITERAL> token
    
        ## for coding-comment: https://stackoverflow.com/questions/25395251/detecting-and-skipping-line-comments-with-flex
        comment = re.sub(r'[/][/].*', r"<SINGLE_LINE_COMMENT>", comment) # replace all single-line-comment with <SINGLE_LINE_COMMENT> token
        comment = re.sub(r'[/][*][^*]*[*]+([^*/][^*]*[*]+)*[/]', r"<MULTY_LINE_COMMENT>", comment) # replace all multy-line-comment with <MULTY_LINE_COMMENT> token
    
#     comment = comment.replace('\u2011', '')
    
    tokenized = tweet_tokenizer.tokenize(comment)
    tokenized = word_punct_tokenizer.tokenize(' '.join(tokenized))
#     print(tokenized)
    tokenized = mwe_tokenizer.tokenize(tokenized)
    tokenized_comment = ' '.join(tokenized[:max_size])
#     print(tokenized)
    tokenized_comment = re.sub(r'[^\x00-\x7f]',r'', tokenized_comment)
    tokenized = tokenized_comment.split()
    return tokenized_comment, tokenized
```

```python
def getCodeToken(abs_tok_truc_file):
    ## https://stackoverflow.com/questions/22216076/unicodedecodeerror-utf8-codec-cant-decode-byte-0xa5-in-position-0-invalid-s
    with open(abs_tok_truc_file,'rb') as f:
#         for line in f:
#             print(line.decode(errors='ignore'))
        tokenized_code = f.readline().decode(errors='replace')
#         if tokenized_code[-1] == '\n':
#             tokenized_code = tokenized_code[:-1]
    tokenized_code = tokenized_code.strip()
    code_tokens = tokenized_code.split()
    return tokenized_code, code_tokens

def getCodeToken_simpler(untokenized_code_file_path, line_number, threshold):
    line_number -= 1
    with open(untokenized_code_file_path, encoding="utf8") as f:
        lines = f.readlines()
    
    min_level = line_number-threshold
    if min_level < 0 :
        min_level = 0
    max_level = line_number+threshold+1
    # code_line_count = len(lines) 
    # if max_level >= code_line_count:
    #     max_level = code_line_count-1
    ## max level count er upore restriction dewa lage na

    top_segment = lines[min_level:line_number]
    line_of_focus = lines[line_number].replace('\n',' ')
    bottom_segment = lines[line_number+1:max_level]
    
    code_segment = ' '.join(top_segment) + '<START_FOCUS>' + line_of_focus + '<END_FOCUS>' + ' '.join(bottom_segment)
    
    _, code_tokens = getCommentToken(code_segment, max_size=MAX_CODE_TOKEN_STREAM_SIZE, code_like_tokenize=True)
    tokezined_code, _ = getCommentToken(code_segment, max_size=MAX_CODE_TOKEN_STREAM_SIZE)
    
    return tokezined_code, code_tokens
```

```python
# def generateAbsTokTrucCode(untokenized_code_file_path, line_number):
#     abs_tok_truc_file = '2_abstract_tokenized_truncated.txt' # set a fixed file, which will be overwritten everytime, if we want to save disk space.
#     # we use different output files for different input files. thought i kept a fixed one: 'abstract_tokenized_truncated.txt'
    
#     ### run the subprocesses to generate the files-> abstract.txt, abstract_tokenized.txt abstract_tokenized_truncated.txt
    
#     return abs_tok_truc_file
import subprocess

def generateAbsTokTrucCode(untokenized_code_file_path, line_number, destination_dir, file_name):
    abs_tok_file = 'abstract_tokenized.txt'
    abs_tok_truc_file = 'abstract_tokenized_truncated.txt' # set a fixed file, which will be overwritten everytime, if we want to save disk space.
    # we use different output files for different input files. thought i kept a fixed one: 'abstract_tokenized_truncated.txt'
    
    ### run the subprocesses to generate the files-> abstract.txt, abstract_tokenized.txt abstract_tokenized_truncated.txt
    ##java -jar F:\sequencer\src\lib\abstraction-1.0-SNAPSHOT-jar-with-dependencies.jar F:\temp\ClassStudent.java 18 F:\temp
    ##python F:\sequencer\src\Buggy_Context_Abstraction\tokenize.py F:\temp\ClassStudent_abstract.java F:\temp\ClassStudent_abstract_tokenized.txt
    ##perl F:\sequencer\src\Buggy_Context_Abstraction\trimCon.pl F:\temp\ClassStudent_abstract_tokenized.txt F:\temp\ClassStudent_abstract_tokenized_truncated.txt 1000
    
    ## the shell commands
    ## HYPERPARAM Tufano
    java_cmd = "java -jar \"E:/Thesis 1505020/sequencer-master/src/lib/abstraction-1.0-SNAPSHOT-jar-with-dependencies.jar\" \"" +  untokenized_code_file_path + "\" " + str(line_number) + " \"" + destination_dir + "\""
    python_cmd = "python \"E:/Thesis 1505020/sequencer-master/src/Buggy_Context_Abstraction/tokenize.py\" \"" + destination_dir + '/' +file_name+ "_abstract.java\" \"" +  destination_dir + "/" + abs_tok_file + "\""
    perl_cmd = "perl \"E:/Thesis 1505020/sequencer-master/src/Buggy_Context_Abstraction/trimCon.pl\" \"" + destination_dir + "/" + abs_tok_file + "\" \"" + destination_dir + "/" + abs_tok_truc_file + "\" " + str(MAX_CODE_TOKEN_COUNT)
    
    # E:\Thesis 1505020\sequencer-master\src\Buggy_Context_Abstraction
    # E:\Thesis 1505020\sequencer-master\src\lib
    
    #print("java_cmd: {}\n\npython_cmd: {}\n\nperl_cmd: {}\n\n".format(java_cmd, python_cmd, perl_cmd))
    
    ## execute above commands in python subprocess
#     print('java:', java_cmd)
    
    check = subprocess.call(java_cmd, shell=True)
    if check != 0:
        print("java_cmd execusion failed")
        return None
    check = subprocess.call(python_cmd, shell=True)
    if check != 0:
        print("python_cmd execusion failed")
        return None
    check = subprocess.call(perl_cmd, shell=True)
    if check != 0:
        print("perl_cmd execusion failed")
        return None
    
    abs_tok_truc_file = destination_dir + "/" + abs_tok_truc_file
    
    abs_file_remove_cmd = "del \"" + destination_dir + "/" +file_name+ "_abstract.java\""
    abs_tok_file_remove_cmd = "del \"" + destination_dir + "/" + abs_tok_file + "\""
    
    #print(abs_file_remove_cmd)
    #print(abs_tok_file_remove_cmd)
    
    check = subprocess.call(abs_file_remove_cmd.replace('/', '\\'), shell=True)
    if check != 0:
        print("abs_file_remove_cmd execusion failed")
        return None
    
    check = subprocess.call(abs_tok_file_remove_cmd.replace('/', '\\'), shell=True)
    if check != 0:
        print("abs_tok_file_remove_cmd execusion failed")
        return None
    
    return abs_tok_truc_file



## HYPERPARAM
# if RUN_SEQUENCER==True: sequenceR er pipeline run hoye preprocess hobe
def getSingleCodeReviewDatapoint(data, basePath, threshold, RUN_SEQUENCER=False):
    untokenized_code_file_path = basePath +'/'+data['code_file_name']+'/'+str(data['base_patch_number'])+'.java'
    destination_dir = r"D:\APR\MARK_1\DATA\TEMP_PREPROCESSED_CODE"
    
    if RUN_SEQUENCER:
        abs_tok_truc_file = generateAbsTokTrucCode(untokenized_code_file_path=untokenized_code_file_path, line_number=data['actual_line_number'], destination_dir=destination_dir , file_name=str(data['base_patch_number']))   
        if abs_tok_truc_file is None:
            return False

        tokenized_code, code_tokens = getCodeToken(abs_tok_truc_file=abs_tok_truc_file)

        ## remove the abs_tok_truc_file as it is no longer required for this iteration
        abs_tok_truc_file_remove_cmd = 'del \"' + abs_tok_truc_file + "\""
        check = subprocess.call(abs_tok_truc_file_remove_cmd.replace('/', '\\'), shell=True)
        if check != 0:
            print("abs_tok_truc_file_remove_cmd execusion failed")
            return False

#     ## generate code tokens in a simpler way
    else:
        tokenized_code, code_tokens = getCodeToken_simpler(untokenized_code_file_path=untokenized_code_file_path, line_number=data['actual_line_number'], threshold=threshold)
    
    ## generate reviewer's comment's tokens
    comment = data['message']
    tokenized_comment, _ = getCommentToken(comment=comment)
    
    inputString = "<START_COMMENT> " + tokenized_comment + " <END_COMMENT> " + "<START_CODE> " + tokenized_code + " <END_CODE>"  
    inputString = inputString.replace('\n', "")

#     print('\n' in inputString)

    _, target_tokens = getCommentToken(comment=data['changed_code'].replace('\n', ""), max_size=MAX_TARGET_TOKEN_STREAM_SIZE, code_like_tokenize=True)
    tokenized_target, _ = getCommentToken(comment=data['changed_code'].replace('\n', ""), max_size=MAX_TARGET_TOKEN_STREAM_SIZE)

    if len(code_tokens) < MIN_code_tok_cnt:
        print('len(code_tokens) < MIN_code_tok_cnt:', code_tokens)
        return False
    if len(comment_tokens) < MIN_cmnt_tok_cnt:
        print('len(comment_tokens) < MIN_cmnt_tok_cnt:', comment_tokens)
        return False
    if len(target_tokens) < MIN_tgt_tok_cnt:
        print('len(target_tokens) < MIN_tgt_tok_cnt:', target_tokens)
        return False
    
    X_list.append(inputString)
    Y_list.append(tokenized_target)

    
    global max_code_tok_cnt
    global max_cmnt_tok_cnt
    global max_tgt_tok_cnt
    
    if len(code_tokens) > max_code_tok_cnt:
        max_code_tok_cnt = len(code_tokens)
    if len(comment_tokens) > max_cmnt_tok_cnt:
        max_cmnt_tok_cnt = len(comment_tokens)
    if len(target_tokens) > max_tgt_tok_cnt:
        max_tgt_tok_cnt = len(target_tokens)
    
    
    for tok in code_tokens:
        if tok in code_token_map:
            code_token_map[tok] += 1
        else:
            code_token_map[tok] = 1

    for tok in target_tokens:
        if tok in code_token_map:
            code_token_map[tok] += 1
        else:
            code_token_map[tok] = 1
            
    for tok in comment_tokens:
        if tok in comment_token_map:
            comment_token_map[tok] += 1
        else:
            comment_token_map[tok] = 1
    
    return True

# import glob
import random

            
##the above codes for generating x, y pair should be run in loop for all the files.
def getAllDatapoints(dataBasePath, jsonBasePath, threshold):
    # here jsonBasePath is a directory containing all the json files containing our expected info's. We can set set a specific naming pattern for it though.
    # using regex in glob to get the file names with .json extention.
    # glob documentation-> https://docs.python.org/3/library/glob.html
    
    ###jsonFilePaths = glob.glob(jsonBasePath+'/'+'*.json')
    
    all_data = getJsonData(JsonFile=jsonBasePath)
    
    SHUFFLE_COUNT = 150

    for i in range(SHUFFLE_COUNT):
        random.shuffle(all_data)
    
    failed_datapoints = []
    successful_datapoints = []
    RANGE_ = len(all_data)
#     RANGE_ = 500
    print('Processing {} datapoints'.format(len(all_data)))
#     for i in range(11274, 11274+1):
    for i in range(RANGE_):
        
#         if i == 200: ## FOR ONLY TEST PURPOSE
#             break
                
        data = all_data[i]
        
#         print('data:',i)
        chk = getSingleCodeReviewDatapoint(data=data, basePath=dataBasePath, threshold=threshold)
        if chk is False:
#             failed_datapoints.append(i)
            failed_datapoints.append(data['comment_id'])
            print('Failed to generate {}th data-point. Skipping..'.format(i))
            continue # cause, why not..
        else:
            successful_datapoints.append(data['comment_id'])
        
            
    print("all data points generated.\nlen(X_list): {}, len(Y_list): {}".format(len(X_list), len(Y_list)))
    assert len(X_list) == len(Y_list), "len(X_list): {}, len(Y_list): {}".format(len(X_list), len(Y_list))
    
    return failed_datapoints, successful_datapoints
```

### DeTokenization

```python
REVERSE_WHITESPACE_DICT = {}
for key in WHITESPACE_DICT:
    value = WHITESPACE_DICT[key]
    REVERSE_WHITESPACE_DICT[value] = key
```

```python
def extreme_detokenization(tokens):
    s = ""
    for token in tokens:
        if token in REVERSE_WHITESPACE_DICT:
            s+= REVERSE_WHITESPACE_DICT[token]
        else:
            s+= token
    return s
```
